<p>Importar librerías</p>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.set_option("display.max_columns", 200)
pd.set_option("display.width", 200)
```

<p>Cargar dataset</p>

```python
df = pd.read_csv("data.csv", sep="\t")

print("Dataset cargado correctamente")
print("Filas y columnas:", df.shape)

df.head()
```

<p>Información general del dataset</p>

```python
df.info()
```

<p>Valores nulos</p>

```python
df.isnull().sum().sort_values(ascending=False).head(25)
```

<p>Identificar columnas por tipo</p>

```python
tipi_cols = [f"TIPI{i}" for i in range(1, 11) if f"TIPI{i}" in df.columns]
dass_cols = [c for c in df.columns if c.startswith("Q") and c.endswith("A")]
vcl_cols  = [c for c in df.columns if c.startswith("VCL")]

print("TIPI:", tipi_cols)
print("Total DASS:", len(dass_cols))
print("Total VCL:", len(vcl_cols))
```

LIMPIEZA 1: Filtrado de edad

```python
df_clean = df.copy()

df_clean = df_clean[(df_clean["age"] >= 13) & (df_clean["age"] <= 100)]

print("Filas después de filtrar edad:", df_clean.shape[0])
```

LIMPIEZA 2: Validación TIPI

```python
for col in tipi_cols:
    df_clean = df_clean[df_clean[col] > 0]

print("Filas después de validar TIPI:", df_clean.shape[0])
```

Definir ítems DASS

```python
dep_items = [3,5,10,13,16,17,21,24,26,31,34,37,38,42]
anx_items = [2,4,7,9,15,19,20,23,25,28,30,36,40,41]
str_items = [1,6,8,11,12,14,18,22,27,29,32,33,35,39]

dep_cols = [f"Q{i}A" for i in dep_items]
anx_cols = [f"Q{i}A" for i in anx_items]
str_cols = [f"Q{i}A" for i in str_items]
```

LIMPIEZA 3: Validación DASS + cálculo de puntajes

```python
dass_cols = dep_cols + anx_cols + str_cols

for col in dass_cols:
    df_clean = df_clean[df_clean[col].between(1,4)]

df_clean["DepressionScore"] = df_clean[dep_cols].sum(axis=1) - len(dep_cols)
df_clean["AnxietyScore"]    = df_clean[anx_cols].sum(axis=1) - len(anx_cols)
df_clean["StressScore"]     = df_clean[str_cols].sum(axis=1) - len(str_cols)

df_clean[["DepressionScore","AnxietyScore","StressScore"]].describe()
```

Categorización

```python
def categorize(score):
    if score <= 13:
        return 0
    elif score <= 20:
        return 1
    else:
        return 2

df_clean["Dep_Level"] = df_clean["DepressionScore"].apply(categorize)
df_clean["Anx_Level"] = df_clean["AnxietyScore"].apply(categorize)
df_clean["Str_Level"] = df_clean["StressScore"].apply(categorize)
```

Visualización de niveles

```python
df_clean["Dep_Level"].value_counts().sort_index().plot(kind="bar", title="Depresión")
plt.show()

df_clean["Anx_Level"].value_counts().sort_index().plot(kind="bar", title="Ansiedad")
plt.show()

df_clean["Str_Level"].value_counts().sort_index().plot(kind="bar", title="Estrés")
plt.show()
```
